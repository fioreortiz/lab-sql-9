```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
data = pd.read_sql_query('SELECT * FROM rental', engine)
data = pd.read_sql_query("SELECT rental_id,rental_date, CASE WHEN DAYOFWEEK(rental_date) IN (1,7) then 'weekend' WHEN DAYOFWEEK(rental_date) IN (2,3,4,5,6) then 'workday' ELSE 'No status' END AS 'day_type' FROM sakila.rental",engine)
data
```

```python
# 4. Get 10 the longest films.
data = pd.read_sql_query("SELECT * FROM sakila.film ORDER BY LENGTH(length) DESC LIMIT 10", engine)
data
```

```python
# 5. How many films include Behind the Scenes content?
data = pd.read_sql_query("SELECT * FROM sakila.film WHERE special_features LIKE '%%Behind the Scenes%%'", engine)
data
```

```python
# 6. Which kind of movies (rating) have a mean duration of more than two hours?
data = pd.read_sql_query("SELECT rating, round(AVG(length),2) AS mean_duration FROM sakila.film GROUP BY rating HAVING mean_duration > 120", engine)
data
```

```python
# 7. Rank films by length (filter out the rows that have nulls or 0s in length column). 
# In your output, only select the columns title, length, and the rank.
data = pd.read_sql_query("SELECT title, length, dense_rank() OVER(ORDER BY (length)) as 'rank' FROM sakila.film WHERE (length IS NOT NULL) OR (length = '  ')", engine)
data
```
